Los datos fueron descargados de la página oficial de la **Red Sismológica Nacional de Colombia (RSNC)**, la cual es una base actual y que permite realizar una búsqueda experta que puede filtrar según múltiples opciones.

En este caso, se descargan los sismos que tengan magnitud mayor que 4.0 debido a la gran cantidad que se han registrado desde 1993, específicamente más de 172.639. 

Por lo tanto se ingresa al link de Consulta experta:

+ http://200.119.88.135/RSNC/index.php/consultas/consulexp
+ En la opción *Rango de Magnitud local(Ml)*, *Magnitud mínima* se coloca 4.0 y luego click en *consultar*.

De esta manera aparece un recuadro donde se permite descargar un formato de excel que viene con el nombre de 'reporte_rsnc_experta.xls'.

Una vez encontrado se cambió el nombre a 'sismos.xls'.

```python
import pandas as pd
```

```python
sismos = pd.read_html('sismos.xls', header=9)
```

```python
sismos = sismos[0]
```
